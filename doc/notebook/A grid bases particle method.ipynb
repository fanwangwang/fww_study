# A grid bases particle method

这篇文章主要由 Shinyu Leung, Hongkai Zhao 提出的关于移动界面的一个新算法:基于网格质点的方法.这个方法结合了拉格朗日和欧拉格式的优点，用拟一致无网格质点来表示和追踪移动界面，避免对于追踪方法重新网格化的困难，基础网格提供了无网格质点的欧拉参考，质点邻域的信息和样本密度，，不需要解 PDE ,这个方法通常很自然地合并适应性和拓扑控制，未来的工作包括开界面移动的扩展，解移动界面的 PDE .

## 介绍

移动界面的模型在科学和工程的许多问题中是很重要的，数值方面，有两个挑战:
1. 带有复杂的几何和动力学的一个多样化的表示和追踪
1. 在移动界面和整体动力学的耦合

这篇文章主要关注于第一个问题并且发展一个基于网格质点的方法来表示和追踪移动界面.

一般来说，可以将现有的移动界面问题的数值方法分为两类：
1. 拉格朗日追踪方法
1. 捕捉方法

拉格朗日追踪方法的优点:
1. 界面的准确表示
1. 追踪质点移动的简单性

拉格朗日追踪方法的缺点:很难保持具有包括大的变形或拓扑变化的复杂几何形状和动力学的移动界面的光滑全局参数化或拟一致网格，特别是在高维度上


捕捉方法的优点：
1. 在具有简单数据结构的固定网格上,几何问题转化成偏微分方程问题（无需参数化或表面网格）
1. 可以在粘度意义上容易地处理诸如合并或分割的拓扑改变

捕捉方法的缺点：
1. 由于隐式表示，捕获方法通常在界面表示和演进两方面都比跟踪方法更不精确和更低效
1. 更多的数值耗散
1. 需要额外的努力来明确地确定界面
1. 同时解决基于笛卡尔网格的PDE使得网格加密和适应性变得更复杂

上述数值方法的另一个困难是如何根据物理控制拓扑变化.拉格朗日跟踪方法可以很容易地模拟界面的运动，因为界面上的质点通过参数化或表面网格局部连接，当界面的两个不同部分在物理空间中接近时，它们不会感觉到彼此，因为在参数空间或在表面网格中它们相距很远，这两个部分独立移动并穿过彼此. 然而，合并或分割是如上所述的拉格朗日跟踪方法的主要困难.相反，欧拉
捕获方法将界面嵌入单值标量函数中. 因此，拓扑变化，如合并或分裂可以容易地处理，而对于大多数捕获方法，界面交叉难以处理

在本文中，提出了一个新的框架来模拟界面运动，以及拉格朗日和欧拉公式的优点. 基本思想是像通常的拉格朗日方法使用拟一致无网格质点来显式地表示和跟踪界面，而基础欧拉网格作为参考那些质点，这里是给出方法的几个关键功能
1. 界面由没有网格或参数化的质点表示
1. 粒子的采样具有对在其附近的基础网格点是一对一参考的
1. 自适应采样的界面可以通过基础网格的局部网格加密轻松实现，因为没有在网格上的求解PDE
1. 拓扑变化可以使用拉格朗日和欧拉信息控制

文章中提到了向前追踪方法，质点水平集方法，混合基础网格结合拉格朗日质点的方法，这些方法不能直接用于高阶运动中
